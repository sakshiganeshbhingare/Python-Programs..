#Dictionary

```python
#Create a Dictionary
student={
    "Name":"Sakshi",
    "age": 20,
    "course":"Python"
}
print(student)
```

```python
#Access Value
student={"name":"Partik","age":18}

print(student["name"])
print(student["age"])
```

```python
#Add New Item
student={"name":"Sakshi"}
student["City"]="Pune"
print(student)
```

```python
#Update Value
student={"name":"Sakshi", "age": 20}
student["age"]=21
print(student)
```

```python
#Delete Item Using(del)
student={"name":"Sakshi", "age": 20}
del student["age"]
print(student)
```

```python
#Remove Item Using(pop)
student={"name":"Sakshi", "age": 20}
student pop["age"]
print(student)
```
