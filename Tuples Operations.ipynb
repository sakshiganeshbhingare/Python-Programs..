#Tuples

```python
t=(10,20,30,40)
print(t)
```

```python
#single element tuple
t=(10,)
print(t)
```

```python
t=10,20,30
print (t)
```

```python
#Accessing Elements
t=(1,2,3,4,5,6)
print(t[0])#Positive index
print(t[2])
```

```python
t=(6,5,4,3,2,1)
print(t[-1])
print(t[-2])
```

```python
#Slicing
t=(10,20,30,40,50)
print(t)
print(t[1:4])
print(t[:3])
print(t[2:])
```

```python
#Concatenation
t1=(1,2,3)
t2=(4,5,6)
print(t1+t2)
```

```python
#Repetion
t=(10,20)
print(t*3)
print(t*2)
```

```python
#Membership Operators
t=(10,20,30)
print(20 in t)
print(50 not in t)
print(60 in t)
```

```python
#Length
t=(10,20,30,40)
print(len(t))
```

```python
#Maximum and Minimum
t=(10,20,30,40,5)
print(max(t))
print(min(t))
```

```python
#Sum
t=(10,20,30)
print(sum(t))
```

```python
#Count
t=(10,20,30,40,50,10,30)
print(t.count(10))
```

```python
#Index
t=(10,20,30,40)
print(t.index(30))
```

```python
#For loop
t=(10,20,30,40)
for i in t:
    print (i)
```

```python
#Nested Tuple
t=(1,2),(3,4),(5,6)
print(t[1])
print(t[2][1])
print(t[1][1])
print(t[-3][1])
```

```python
#Packing
t=10,20,30,40
print(t)
```

```python
#Unpacking
t=(10,20,30,40)
a,b,c,d=t
print(a)
print(b)
print(c)
print(d)
```

```python
#Convert list to Tuple
l=[10,20,30,40]
t=tuple(l)
print(t)
```

```python
#Convert Tuple to List
t=(10,20,30,40)
l=list(t)
print(l)
```

```python
#Reverse Using Slicing
t=(10,20,30,40)
print(t[::-1])
```

```python
#Immutability
t=(10,20,30)
t[1]=100
```
