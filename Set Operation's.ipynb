```python
#WAP to genarate List of randum Numbers and make to differnd odd number list or even numbers list
L1=[4,6,5,3,2,1,7,9,8,10]
OL=[]
EL=[]
for i in L1:
    if i%2==0:
        EL.append(i)
    else:
        OL.append(i)
print(f"Original list : {L1}\nEven number list : {EL}\nOdd number list : {OL}")

    
```
