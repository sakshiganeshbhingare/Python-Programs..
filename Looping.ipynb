```python
#Check Even Or Odd Number(if-else)
num=int(input("Enter the number:"))

if num%2==0:
    print("Even Number")
else:
    print("Odd Number")
```

```python
#Find Largest of Two Numbers(if-else)
a=int(input("Enter the First Number:"))
b=int(input("Enter the Second Number:"))

if a>b:
    print("Largest=", a)
else:
    print("Largest=", b)
```

```python
#if-elif-else
num=int(input("Enter the First Number:"))
num=int(input("Enter the Second Number:"))
num=int(input("Enter the Third Number:"))

if num>0:
    print("Largest=", num)
elif num<0:
    print("Largest=", num)
else:
    print("Largest=", num)
```

```python
#Check Positive,Negative, or Zero
num=int(input("Enter the Number:"))

if num>0:
    print("Positive Number")
elif num<0:
    print("Negative Number")
else:
    print("Zero")
```

```python
#Looping(For loop)
#Print the Numbers From 1 t0 10
for i in range(1,11):
    print(i)
```

```python
#Print Even Numbers From 1 to 20
for i in range(2,21,2):
    print(i)
```

```python
#Sum of Numbers from 1 to 100
total=0

for i in range(1,101):
    total =total+i
print("Sum", total)
```

```python
#Print Multipication Table
num=int(input("Enter the Number"))

for i in range(1,11):
    print(num, "X", i, "=", num *i)
```

```python
#While Loop
i=1

while i<=10:
    print(i)
    i=i+i
```

```python
#Break Statement
for i in range(1,11):
    if i ==6:
        break
    print(i)
```

```python
#Continue Statement
for i in range(1,11):
    if i ==4:
        continue
    print(i)
```

```python
#Pass()
for i in range(1,6):
    if i == 3:
        pass
    print(i)
```
